```python

#methode1 de classification
import re

# Liste de mots-clés classés par domaine
mots_cles_domaine = {
    'IT': ['Software Engineer', 'IT Specialist', 'Network Administrator', 'System Analyst', 'Data Scientist',
           'Data Analyst', 'Full Stack Developer', 'Developer', 'Front End Developer', 'Back End Developer',
           'DevOps Engineer', 'Cybersecurity Specialist'],
    'FINANCE': ['Finance Manager', 'Accountant', 'Financial Analyst', 'Budget Analyst', 'Auditor',
                'Tax Advisor', 'Credit Analyst', 'Investment Analyst', 'Risk Manager', 'Treasury'],
     'MARKETING': ['Marketing Coordinator', 'Digital Marketing Specialist', 'Social Media Manager',
                  'Content Strategist', 'SEO Specialist', 'Brand Manager', 'Marketing Analyst',
                  'Public Relations Specialist', 'Advertising Manager', 'Product Marketing Manager']
    # Ajoutez plus de domaines et de mots-clés ici selon vos besoins
}

def classifier_description(description):
    # Convertir en minuscules pour simplifier la correspondance
    description = description.lower()

    for domaine, mots_cles in mots_cles_domaine.items():
        for mot_cle in mots_cles:
            # Convertir le mot-clé en expression régulière et vérifier la correspondance
            if re.search(r'\b' + re.escape(mot_cle.lower()) + r'\b', description):
                return domaine  # Retourner le domaine correspondant

    return 'Unknown'  # Retourner 'Unknown' si aucun mot-clé ne correspond

# Exemple d'utilisation avec un DataFrame pandas
# Assurez-vous que df_usa est déjà défini et contient la colonne 'job_title'
# df_usa = pd.DataFrame({'job_title': ['Software Engineer', 'Accountant', 'Digital Marketing Specialist', 'Unknown Job']})

# Appliquer la fonction classifier_description à la colonne 'job_title'
df_usa['category'] = df_usa['job_title'].apply(classifier_description)

# Afficher le comptage des catégories trouvées
print(df_usa['category'].value_counts())
```
