```python
keywords_list = {
    "IT": [
        "Software Engineer",
        "Web Developer",
        "Mobile Developer",
        "Frontend Developer",
        "Backend Developer",
        "Full Stack Developer",
        "Game Developer",
        "Application Developer",
        "Embedded Systems Developer",
        "QA Engineer",
        "Software Tester",
        "DevOps Engineer",
        "IT Specialist",
        "Network Administrator",
        "System Analyst",
        "Data Scientist",
        "Data Analyst",
        "Developer",
        "Front End Developer",
        "Back End Developer",
        "Cybersecurity Specialist",
        "Big Data Engineer",
        "Machine Learning Engineer",
        "AI Specialist",
        "Business Intelligence (BI) Analyst",
        "Data Architect",
        "Statistician",
        "Data Visualization Specialist",
        "Cybersecurity Analyst",
        "Information Security Analyst",
        "Security Engineer",
        "Ethical Hacker",
        "SOC Analyst",
        "Security Consultant",
        "Cyber Threat Analyst",
        "Cryptographer",
        "Incident Responder",
        "Cyber Forensics Analyst",
        "Network Engineer",
        "Network Architect",
        "Telecommunications Engineer",
        "VoIP Engineer",
        "Wireless Engineer",
        "Network Security Engineer",
        "Network Support Specialist",
        "Network Operations Center (NOC) Engineer",
        "System Administrator",
        "Linux Administrator",
        "Windows Administrator",
        "IT Support",
        "IT Technician",
        "Virtualization Engineer",
        "Database Administrator (DBA)",
        "DBA",
        "Server Administrator",
        "Cloud Administrator",
        "software engineer",
        "it specialist",
        "system analyst",
        "cybersecurity specialist",
        "data",
        "it",
        "analyst"
    ],
    "Finance": [
        "Finance Manager",
        "Accountant",
        "Financial Analyst",
        "Budget Analyst",
        "Auditor",
        "Tax Advisor",
        "Credit Analyst",
        "Investment Analyst",
        "Risk Manager",
        "Treasury",
        "financial"
    ],
    "Marketing": [
        "Marketing Coordinator",
        "Digital Marketing Specialist",
        "Social Media Manager",
        "Content Strategist",
        "SEO Specialist",
        "Brand Manager",
        "Marketing Analyst",
        "Public Relations Specialist",
        "Advertising Manager",
        "Product Marketing Manager",
        "Paid Social",
        "marketing",
    ],
    "Sales": [
        "Sales Representative",
        "Account Executive",
        "Sales Manager",
        "Business Development Manager",
        "Sales Analyst",
        "Account Manager",
        "Sales Director",
        "Sales Operations Specialist",
        "Regional Sales Manager",
        "Inside Sales Representative",
        "sales",
        "cashier",
        "retail",
        "store",
        "salesman"
    ],
    "Healthcare": [
        "registered nurse",
        "medical director",
        "clinical research coordinator",
        "pediatric",
        "health",
        "healthcare",
        "nurse",
        "travel rn",
        "Hospice"
    ],
    "Education": ["school principal", "college counselor", "special education teacher", "school", "teacher", "Instructor"],
    "Engineering": ["civil engineer", "mechanical engineer", "electrical engineer", "Manufacturing Engineer"],
    "Laws": ["family lawyer", "criminal defense attorney", "legal secretary", "law"],
    "Real Estate": ["real estate agent", "property manager", "real estate appraiser"],
    "Food": ["cook", "cuisine", "restaurant", "cheese"],
    "Construction": ["construction"],
    "HR": ["hr", "human resources"],
    "Textile": ["Sewer", "textile"]
}

for domain, keywords in keywords_list.items():
  keywords_list[domain] = [keyword.lower() for keyword in keywords]

def categorize(desc):
  desc = desc.lower()

  for domain, keywords in keywords_list.items():
    if any(keyword in desc for keyword in keywords):
      return domain

  return 'Others'


df['category'] = df['job_title'].apply(categorize)
df['category']
```
